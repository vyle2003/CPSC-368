```python
import pandas as pd
import altair as alt
import matplotlib.pyplot as plt
```

```python
USAdata = pd.read_csv("/home/jupyter/CPSC_368/finalpaper/data/usa_deaths_data.csv")
USAdata
```

```python
CAdata = pd.read_csv("/home/jupyter/CPSC_368/finalpaper/data/canada_deaths_data.csv")
CAdata
```

```python
combined = pd.concat([USAdata, CAdata], ignore_index=True)
combined
```

```python
combined['YEAR'] = combined['YEAR'].astype(str)
```

## 1. How did the age distribution of COVID-19 deaths differ between Canada and the USA over time?

```python
age_line = alt.Chart(combined).mark_line(point=True).encode(
    x='YEAR:N',
    y='DEATHS:Q',
    color='COUNTRY:N',
    tooltip=['YEAR', 'COUNTRY', 'AGE_GROUP', 'DEATHS']).facet(
    facet='AGE_GROUP:N',
    columns=3).resolve_scale(
    y='independent').properties(
    title='COVID-19 Deaths Over Time by Age Group (Canada vs USA)')

age_line
```

##  2. Were there specific age groups that experienced disproportionately higher mortality rates in one country compared to the other?

```python
age_bar = alt.Chart(combined).mark_bar().encode(
    x=alt.X('AGE_GROUP:N', title='Age Group'),
    y='DEATHS:Q',
    color='COUNTRY:N',
    column=alt.Column('YEAR:N', title='Year'),
    tooltip=['YEAR', 'AGE_GROUP', 'COUNTRY', 'DEATHS']).resolve_scale(
    x='independent').properties(
    title='Age-Specific COVID-19 Deaths per Country by Year')

age_bar
```

## 3. Is there any correlation between age-group and mortality rate from COVID-19 in Canada and the United States?

```python
bubble_chart = alt.Chart(combined).mark_circle().encode(
    x=alt.X('AGE_GROUP:N', title='Age Group'),
    y=alt.Y('COUNTRY:N', title='Country'),
    size=alt.Size('DEATHS:Q', scale=alt.Scale(range=[0, 3000]), title='Number of Deaths'),
    color=alt.Color('GENDER:N', title='Gender'),
    tooltip=['YEAR', 'AGE_GROUP', 'GENDER', 'COUNTRY', 'DEATHS']).properties(
    width=200,
    height=200).facet(
    facet=alt.Facet('YEAR:N', title='Year'),
    columns=2).resolve_scale(
    size='independent').properties(
    title='COVID-19 Mortality by Age, Country, and Gender')

bubble_chart
```

## 4. How did the pandemic impact mortality rates differently for males and females across various age groups in Canada and the United States?

```python
stacked = alt.Chart(combined).mark_bar().encode(
    y=alt.Y('AGE_GROUP:N', title='Age Group'),
    x=alt.X('DEATHS:Q', stack='zero', title='COVID-19 Deaths'),
    color=alt.Color('GENDER:N', title='Gender'),
    tooltip=['YEAR', 'COUNTRY', 'AGE_GROUP', 'GENDER', 'DEATHS']
).facet(column='YEAR:N',row='COUNTRY:N').resolve_scale(x='independent').properties(
    title='COVID-19 Deaths by Gender, Age Group, Year, and Country')

stacked
```
